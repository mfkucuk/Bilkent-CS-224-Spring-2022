```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()  
ax.set_title('Miss Rate vs. Block Size (50x50)')
ax.set_xlabel('Block Size')
ax.set_ylabel('Miss Rate')

ax.plot([16, 32, 64, 128, 256],
        [0.06, 0.03, 0.06, 0.05, 0.01], 
        label='2048 bytes', marker='o')

ax.plot([16, 32, 64, 128, 256],
        [0.06, 0.03, 0.02, 0.01, 0.1], 
        label='4096 bytes', marker='x')

ax.plot([16, 32, 64, 128, 256],
        [0.06, 0.03, 0.02, 0.01, 0], 
        label='8192 bytes', marker='^')

ax.plot([16, 32, 64, 128, 256],
        [0.06, 0.03, 0.02, 0.01, 0], 
        label='16384 bytes', marker='s')

ax.plot([16, 32, 64, 128, 256],
        [0.06, 0.03, 0.02, 0.01, 0], 
        label='32768 bytes', marker='*')

ax.legend()
```
